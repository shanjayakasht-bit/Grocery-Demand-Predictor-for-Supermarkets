```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('data/grocery_sales.csv')
```

```python
df.head()
```

```python
print(df.info())
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['date'], df['sales'])
plt.title("Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
sns.scatterplot(x='temperature', y='sales', data=df)
plt.title("Temperature vs Sales")
plt.show()
```

```python
sns.barplot(x='holiday', y='sales', data=df)
plt.title("Holiday vs Sales")
plt.show()
```
